```python
import pandas as pd 
```

```python
df=pd.read_csv("C:\\Users\\Document\\OneDrive\\Documents\\csv files\\salaries.csv")
df  
```

```python
new_df=df.drop('salary_more_then_100k',axis=1) 
new_df 
```

```python
target=df['salary_more_then_100k'] 
target 
```

```python
from sklearn.preprocessing import LabelEncoder 
le_new_df=LabelEncoder()
new_df['company_n']=le_new_df.fit_transform(new_df['company']) 
new_df['job_n']=le_new_df.fit_transform(new_df['job'])
new_df['degree_n']=le_new_df.fit_transform(new_df['degree']) 
```

```python
new_df 
```

```python
df_me =new_df.drop(['company','job','degree'],axis=1) 
df_me 
```

```python
from sklearn import tree 
model=tree.DecisionTreeClassifier() 
model.fit(df_me,target) 
```

```python
model.score(df_me,target) 
```

```python
model.predict([[2,1,1]])  
```
